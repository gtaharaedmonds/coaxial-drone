```python
from runner import test_run, TestPlan, TestStep
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Create a test plan consisting of steps.
step_duration_ms = 5_000
step_range = [0,0]

steps = [
    TestStep(bottom_throttle=i, top_throttle=i, duration_ms=step_duration_ms)
    for i in step_range
]
plan = TestPlan("test", steps=steps)

plan
```

```python
test_directory = Path("../test_data/load_cell_calibration_data/")
```

```python
data_dict = {}
data_dict["test_load_cell_offset"] = test_run(
    filename=test_directory / "test_load_cell_offset.csv", plan=plan
)
data_dict["test_load_cell_200g"] = test_run(
    filename=test_directory / "test_load_cell_200g.csv", plan=plan
)
data_dict["test_load_cell_200g_0"] = test_run(
    filename=test_directory / "test_load_cell_200g_0.csv", plan=plan
)
data_dict["test_load_cell_200g_1"] = test_run(
    filename=test_directory / "test_load_cell_200g_1.csv", plan=plan
)
data_dict["test_load_cell_1kg"] = test_run(
    filename=test_directory / "test_load_cell_1kg.csv", plan=plan
)
data_dict["test_load_cell_1kg_0"] = test_run(
    filename=test_directory / "test_load_cell_1kg_0.csv", plan=plan
)
data_dict["test_load_cell_1kg_1"] = test_run(
    filename=test_directory / "test_load_cell_1kg_1.csv", plan=plan
)
data_dict["test_load_cell_1200g"] = test_run(
    filename=test_directory / "test_load_cell_1200g.csv", plan=plan
)
data_dict["test_load_cell_1200g_0"] = test_run(
    filename=test_directory / "test_load_cell_1200g_0.csv", plan=plan
)
data_dict["test_load_cell_1200g_1"] = test_run(
    filename=test_directory / "test_load_cell_1200g_1.csv", plan=plan
)
data_dict["test_load_cell_1500g"] = test_run(
    filename=test_directory / "test_load_cell_1500g.csv", plan=plan
)
data_dict["test_load_cell_1500g_0"] = test_run(
    filename=test_directory / "test_load_cell_1500g_0.csv", plan=plan
)
data_dict["test_load_cell_1500g_1"] = test_run(
    filename=test_directory / "test_load_cell_1500g_1.csv", plan=plan
)
data_dict["test_load_cell_2kg"] = test_run(
    filename=test_directory / "test_load_cell_2kg.csv", plan=plan
)
data_dict["test_load_cell_2kg_0"] = test_run(
    filename=test_directory / "test_load_cell_2kg_0.csv", plan=plan
)
data_dict["test_load_cell_2kg_1"] = test_run(
    filename=test_directory / "test_load_cell_2kg_1.csv", plan=plan
)
data_dict["test_load_cell_2kg_2"] = test_run(
    filename=test_directory / "test_load_cell_2kg_2.csv", plan=plan
)
```

```python
# Aggregate test data
df_all = pd.concat(data_dict, names=["test_name"])
df_all.loc['test_load_cell_offset']
```

```python
# Generate buckets for averaging based on step duration defined in testing
stable_step_duration_ms = step_duration_ms / 2
transient_offset_ms = (
    500  # Ignore first 500 ms of each throttle step to allow transients to settle
)
step_start_times = np.arange(len(step_range)) * step_duration_ms


# Aggregate data for each throttle step
df_steps = (
    pd.concat(
        [
            df_all.loc[
                (
                    slice(None),
                    slice(
                        t_start + transient_offset_ms,
                        t_start + transient_offset_ms + stable_step_duration_ms,
                    ),
                ),
                :,
            ]
            for t_start in step_start_times
        ],
        keys=step_range,
        names=["throttle"],
    )
    .groupby(["test_name", "throttle"])
    .agg(["mean", "std"])
    .sort_values(by=('thrust_N', 'mean'))
)

df_steps
```

```python
hx711_scale_factor = 21400
load_cell_offset_N = df_steps.loc['test_load_cell_offset']['thrust_N']['mean'][0]
load_cell_offset_bin = int(load_cell_offset_N * hx711_scale_factor)

# Remove offset test from df and subtract offset from existing measurements
df_steps_nonzero = df_steps.drop('test_load_cell_offset')
thrust_calibration_series = df_steps_nonzero['thrust_N']['mean'] - load_cell_offset_N
```

```python
mass_true_kg = np.array([0.200, 0.200, 0.200, 1.00, 1.00, 1.00, 1.200, 1.200, 1.200, 1.500, 1.500, 1.500, 2.00, 2.00, 2.00, 2.00])
thrust_true_N = 9.81 * mass_true_kg
```

```python
def calibrated_thrust_N(measured_thrust, gain, offset):
    return gain * measured_thrust + offset
```

```python
from scipy.optimize import curve_fit
```

```python
thrust_fit_popt,thrust_fit_pcov = curve_fit(calibrated_thrust_N, thrust_calibration_series, thrust_true_N, sigma=df_steps_nonzero['thrust_N']['std'], absolute_sigma=True)
thrust_cal_fcn = np.vectorize(lambda T: calibrated_thrust_N(T, thrust_fit_popt[0], thrust_fit_popt[1]))
thrust_line_pts = [5,25]

print(thrust_fit_popt)
```

```python
new_load_cell_gain = 21400 / 1.17
np.round(new_load_cell_gain)
```

```python
plt.scatter(thrust_calibration_series, thrust_true_N, label='Raw Load Cell Data', s=12)
plt.plot(thrust_line_pts, thrust_cal_fcn(thrust_line_pts), color='r', label='Linear Calibration')
plt.ylabel('Loaded Weight (N)')
plt.xlabel('Measured Thrust (N)')
plt.legend()
plt.grid()
```
